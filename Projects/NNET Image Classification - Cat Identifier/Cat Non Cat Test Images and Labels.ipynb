## Cat Non Cat Test Images and Labels

```python
import matplotlib.pyplot as plt
import h5py


f = h5py.File('test_catvnoncat.h5', 'r')


print(list(f.keys()))

# Get the data

dx = f['test_set_x']
dy = f['test_set_y']

fig, axs = plt.subplots(5,10, figsize=(20, 16), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.001)

axs = axs.ravel()

  
for i in range(50):
    axs[i].imshow(dx[i])
    title = "Image Label: " + str(dy[i])
    axs[i].set_title(title)
```
